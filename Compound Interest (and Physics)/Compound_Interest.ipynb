# Compound Interest

This code shows the convergence of the discrete compounding interest formula $(1+r/n)^{tn}$ and its continuum limit $e^{rt}$. 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['figure.dpi'] = 150 

def Compound(n,r,t):
    return (1 + r/n)**(t*n)


def Exponential(r,t):
    return np.exp(r*t)

t_end = 5
n = [1, 12, 52, 365, 1000]
labels = ['Yearly', 'Monthly', 'Weekly', 'Daily', '~Continously']
r = 0.5
t = np.linspace(0,t_end, 1000)

plt.figure()
for i in range(len(n)):
    t_disc = np.arange(0, t_end, 1/n[i])
    plt.step(t_disc, Compound(n[i],r,t_disc),where='post', color = (np.random.randint(0,255,1)[0]/255, np.random.randint(0,255,1)[0]/255, np.random.randint(0,255,1)[0]/255), label = f'{labels[i]} = {n[i]}')

plt.plot(t, Exponential(r,t),'k--', label = 'Exponential')
plt.xlabel("Time (yrs)")
plt.ylabel("Value of investment (unitless)")
plt.legend()
plt.show()
```
